```python
N = 9

backtrack_calls = 0
backtrack_fails = 0

# Check if valid
def is_valid(board, row, col, num):

    for i in range(N):
        if board[row][i] == num or board[i][col] == num:
            return False

    start_row = row - row % 3
    start_col = col - col % 3

    for i in range(3):
        for j in range(3):
            if board[start_row + i][start_col + j] == num:
                return False

    return True


# Forward Checking
def forward_check(board):

    for row in range(N):
        for col in range(N):

            if board[row][col] == 0:

                possible = False

                for num in range(1, 10):
                    if is_valid(board, row, col, num):
                        possible = True
                        break

                if not possible:
                    return False

    return True


# Backtracking Solver
def solve_sudoku(board):
    global backtrack_calls, backtrack_fails

    backtrack_calls += 1

    for row in range(N):
        for col in range(N):

            if board[row][col] == 0:

                for num in range(1, 10):

                    if is_valid(board, row, col, num):

                        board[row][col] = num

                        if forward_check(board) and solve_sudoku(board):
                            return True

                        board[row][col] = 0  # undo

                backtrack_fails += 1
                return False

    return True


# Read board from file
def read_board(filename):
    board = []

    with open(filename, "r") as file:
        for line in file:
            row = [int(ch) for ch in line.strip()]
            board.append(row)

    return board


# Print board
def print_board(board):
    for row in board:
        print(" ".join(str(num) for num in row))


# Main
files = ["easy.txt", "medium.txt", "hard.txt", "veryhard.txt"]

for file in files:

    print("\n===== Solving", file, "=====")

    board = read_board(file)

    backtrack_calls = 0
    backtrack_fails = 0

    if solve_sudoku(board):
        print_board(board)
    else:
        print("No solution found")

    print("Backtrack Calls:", backtrack_calls)
    print("Backtrack Failures:", backtrack_fails)
```
